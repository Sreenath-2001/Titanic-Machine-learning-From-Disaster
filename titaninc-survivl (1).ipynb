### 1. File To CSV

```python
train_data=pd.read_csv('/kaggle/input/titanic/train.csv')
test_data=pd.read_csv("/kaggle/input/titanic/test.csv")
gender_submmision=pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
```

### 2.Preparing Train Data

```python
train_data.describe()
```

```python
train_data.head()
```

```python
train_data['Age']=train_data['Age'].fillna(train_data['Age'].median())
```

```python
train_data.describe()
```

### 3.Mapping 

```python
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
```

```python
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
train_data.describe()
```

```python
train_data['Embarked'].value_counts().idxmax()
```

```python
train_data['Embarked']=train_data['Embarked'].fillna(0)
```

```python
train_data.describe()
```

### 4. X=

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X=train_data[features]
y=train_data.Survived
```

### 5.Specifying and Fitting Model

```python
from sklearn.tree import DecisionTreeRegressor
from  sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
X_train,X_test,y_train,y_test=train_test_split(X,y)
model=DecisionTreeRegressor()

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
mae= mean_absolute_error(y_pred,y_test)

```

### 6.Determining Mean Absolute Error

```python
mae
```

### 7.Preparing Test Data

```python
test_data.head()
```

```python
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})
test_data['Embarked'] = test_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
test_data.describe()
```

```python
test_data['Age']=test_data['Age'].fillna(test_data['Age'].mean())
test_data['Fare']=test_data['Fare'].fillna(test_data['Fare'].mean())
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
test=test_data[features]
```

### 8.Predicting Survivors

```python
y_test_pred=model.predict(test)
```

### 9. Output To CSV

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_test_pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
